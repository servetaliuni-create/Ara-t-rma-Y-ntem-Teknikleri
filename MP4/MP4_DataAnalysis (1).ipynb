```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.mp3.csv')

print("Veriler:")
print(df.head())

# Istatistikler
print("\n--- Istatistikler ---")
print(df.describe())

# Bos veri var mi kontrol
print("\n--- Bos Veri Kontrolu ---")
print(df.isnull().sum())

# Korelasyon tablosu
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

print("\n--- Korelasyon ---")
print(correlation_matrix)

# Grafikler
plt.figure(figsize=(8, 5))
sns.histplot(df['organik_erisim'], kde=True, color='skyblue')
plt.title('Erisim Dagilimi')
plt.xlabel('Erisim Sayisi')
plt.ylabel('Frekans')
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x='kaydetme_sayisi', y='organik_erisim', hue='format', data=df, s=100)
plt.title('Kaydetme ve Erisim Iliskisi')
plt.xlabel('Kaydetme')
plt.ylabel('Erisim')
plt.legend(title='Format')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Reds', fmt=".2f")
plt.title('Korelasyon Haritası')
plt.show()
```

### Sonuçlar

1.  **Veri Durumu:** Verilerde eksik yok. Erişim sayıları gönderi türüne göre değişiyor.
2.  **İlişki:** Tabloya baktığımda kaydetme sayısı ile erişim arasında bayağı yüksek bir ilişki çıktı. Beğeni sayısı o kadar etkilemiyor.
3.  **Grafikler:** Grafiklerde de kaydetme sayısı arttıkça erişimin arttığı belli oluyor.
